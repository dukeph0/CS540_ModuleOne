## Methods and Attributes between Series and DataFrames

- A **DataFrame** is a 2-dimensional table consisting of rows and columns
- Pandas uses a `NaN` designation for cells that have a missing value. NaN is short for "not a number" which will produce `NaN` values with most operations run against them.
- Like with **Series**, Pandas assigns an index position/label to each **DataFrame** row.
- The `hasnans` attribute exists only in a **Series**. The `columns` attribute exists only on a **DataFrame**.
- Some methods/attributes will return different types of data.
- The `info` method returns a summary of the pandas object.

### Pandas
Pandas is usually very good at figuring out which options to use based on the input data, so we often won't need to add arguments to the call; however, there are many options available should we need them, some of which include the following:

| Parameter | Purpose |
| --- | --- |
| `sep` | Specifies the delimiter |
| `header` | Row number where the column names are located; the default option has `pandas` infer whether they are present |
| `names` | List of column names to use as the header |
| `index_col` | Column to use as the index |
| `usecols` | Specifies which columns to read in |
| `dtype` | Specifies data types for the columns | 
| `converters` | Specifies functions for converting data in certain columns |
| `skiprows` | Rows to skip |
| `nrows` | Number of rows to read at a time (combine with `skiprows` to read a file bit by bit) |
| `parse_dates` | Automatically parse columns containing dates into datetime objects |
| `chunksize` | For reading the file in chunks |
| `compression` | For reading in compressed files without extracting beforehand |
| `encoding` | Specifies the file encoding |

## Examining the Data

### Shape
For a **DataFrame**, the `shape` attribute returns a tuple with two elements: the first element represents the number of rows, and the second element represents the number of columns. For example, if a DataFrame has 10 rows and 5 columns, the shape attribute will return (10, 5).

For a **Series**, the `shape` attribute returns a tuple with one element: the number of elements in the Series. For example, if a Series has 7 elements, the shape attribute will return (7,). 

```python
raw_data.shape
```

### Head
The `head()` function in pandas is used to display the first n rows of a DataFrame or Series. If no argument is passed, it defaults to showing the first 5 rows. It's a useful tool for quickly inspecting the data and understanding its structure, especially when dealing with large datasets.

```python
raw_data.head(10)
```

### Tail
The `tail()` method in pandas is used to retrieve the last n rows of a DataFrame or Series. By default, it returns the last 5 rows, but this number can be customized by passing an integer argument n to the method. 

```python
raw_data.tail(10)
```

### Info
The DataFrames object has a method called `info()`, that gives you more information about the data set which can help pinpoint missing values and wrong data types.

```python
raw_data.info()
```


### Columns
Columns in a Pandas DataFrame are similar to columns in a spreadsheet or SQL table. Each column is a Series object, which is a one-dimensional array with an associated index. Data types within a column must be homogenous.

- **Selecting a column:** Columns can be selected using bracket notation `df['column_name']` or dot notation `df.column_name`. The former is more versatile, especially when column names contain spaces or special characters.
- **Adding a column:** New columns can be added by assignment: `df['new_column'] = values`.
- **Deleting a column:** Columns can be deleted using `del df['column_name']` or `df.drop('column_name', axis=1, inplace=True)`.
- **Renaming columns:** Columns can be renamed using `df.rename(columns={'old_name': 'new_name'}, inplace=True)`.
- **Getting column names:** The column names can be accessed as an **Index** object using `df.columns`.
- **Iterating through columns:** You can iterate through the columns using a for loop: `for column in 'df.columns`.
- **Data types of columns:** Each column can have a different data type. The data type of a column can be accessed using `df['column_name'].dtype`. Common data types include int, float, object, datetime, and bool. 
- **Applying functions to columns:** Functions can be applied to columns using the `apply()` method.
- **Calculating between columns:** New columns can be calculated using basic arithmetic operators. For example `df['new_column'] = df['column_1'] + df['column_2']`.

```python
raw_data.columns
```

### Describe
The `describe()` method returns description of the data in the DataFrame.

If the DataFrame contains numerical data, the description contains these information for each column:

    count - The number of not-empty values.
    mean - The average (mean) value.
    std - The standard deviation.
    min - the minimum value.
    25% - The 25% percentile*.
    50% - The 50% percentile*.
    75% - The 75% percentile*.
    max - the maximum value.

***Percentile meaning:** how many of the values are less than the given percentile.*

```python
raw_data.describe()
```
